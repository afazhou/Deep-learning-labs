# Part 3: Reverse Mode Automatic Differentiation

Dynamic Reverse mode AD can be implemented by declaring a class to represent a **value** and the **child** expressions that the value depends on. We've provided the implementation that was shown in the lecture slides as a basis below, but it's missing some parts that will make it useful.

- Can you also implement division (`__truediv__`), subtraction (`__sub__`) and power (`__pow__`)?

```python
import math

class Var:
    def __init__(self, value):
        self.value = value
        self.children = [] #the list of lists/children, a child is a list made of 2 elements: the weight and child expression
        self.grad_value = None #adjoint variables

    def grad(self): #adjoint variables
        if self.grad_value is None:
            self.grad_value = sum(weight * var.grad() #weight=∂child/∂self.value
                                  for weight, var in self.children)
        return self.grad_value #the gradient=sum(weight × the gradient/adjoint variable of the child i.e. 'var.grad()')--chain rule
    
    def __str__(self):
        return str(self.value)

    def __mul__(self, other):# 'self','other' are instances as input
        z = Var(self.value * other.value) #instance of the product
        self.children.append((other.value, z)) # weight=other.value=∂z/∂self.value, 'z' is the child of 'self'('z' depends on 'self')
        other.children.append((self.value, z)) # weight=self.value=∂z/∂other.value, 'z' is the child of 'other'('z' depends on 'other')
        return z # 'z' is an instance as output

    def __add__(self, other):
        z = Var(self.value + other.value) #instance of the sum
        self.children.append((1, z)) # weight=1=∂z/∂self.value, 'z' is the child of 'self'('z' depends on 'self')
        other.children.append((1, z)) # weight=1=∂z/∂other.value, 'z' is the child of 'other'('z' depends on 'other')
        return z

    def __truediv__(self, other):
        z = Var(self.value / other.value) 
        self.children.append((1/other.value, z)) 
        other.children.append((-self.value*other.value**-2, z)) 
        return z  

    def __sub__(self, other):
        z = Var(self.value - other.value) 
        self.children.append((1, z)) 
        other.children.append((-1, z)) 
        return z     

    def __pow__(self, other):
        z = Var(self.value ** other.value) 
        self.children.append((other.value*self.value**other.value-1, z)) 
        other.children.append((math.log(self.value)*self.value ** other.value, z)) 
        return z     
```

## Implementing math functions

Just like when we were looking at Forward Mode AD, we also need to implement some core math functions. Here's the sine function for a `Var`:

```python
def sin(x): #x is an instance as input, z is an instance as output
    z = Var(math.sin(x.value))
    x.children.append((math.cos(x.value), z))
    return z
```

__Task:__ can you implement the _cosine_ (`cos`), _tangent_ (`tan`), and _exponential_ (`exp`) functions in the code block below?

```python
# TODO: implement additional math functions on dual numbers

def cos(x):
    z = Var(math.cos(x.value))
    x.children.append((-math.sin(x.value), z))
    return z

def tan(x):
    z = Var(math.tan(x.value))
    x.children.append((1/math.cos(x.value)**2, z))
    return z

def exp(x):
    z = Var(math.exp(x.value))
    x.children.append((math.exp(x.value), z))
    return z
```

```python
# Tests
assert cos(Var(0)).value == 1
assert tan(Var(0)).value == 0
assert exp(Var(0)).value == 1
```
